## Importing Data

Importing  a text from Wikipedia utilizing Beautiful Soup and urllib  
The text is about President Macky Sall ,the current President of Senegal

```python
import bs4 as bs
import urllib

# link to the wikipedia of the president Senegal
raw_html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Macky_Sall')  
raw_html = raw_html.read()

wiki_html = bs.BeautifulSoup(raw_html, 'lxml')
article_paragraphs = wiki_html.find_all('p')
article_text = ''
for para in article_paragraphs:  
    article_text += para.text
```

```python
article_text
```

# Tokenize

Splitting the data to create a vocabulary, this is done using the with spaces as distinction between each word.

```python
article_text.split()
```

```python
tk_sequence=article_text.split()
vocabulary=sorted(set(tk_sequence))
print(vocabulary)
```

# USE OF MODAL VERBS

```python
freqdist=nltk.FreqDist(w.lower() for w in vocabulary)
freqdist
```

```python
modals=['when','where','what','can','may','could','might','must','will']
print('The number of occurances of the words are: ')
for m in modals:
    print(m+ ':',freqdist[m])
```

```python
otherwords=['Macky','system','Abdoulaye','COVID-19','support']
print('Other Occurances are ')
for i in otherwords:
    print(i +':',freqdist[i])
```

# NLTK REGEXPTOKENIZER

```python
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer(r'\w+|$[0-9.]+\S+')
vocab= tokenizer.tokenize(article_text)
vocab
```

# Stemming

```python
# Porter stemmer
from nltk.stem.porter import PorterStemmer
stemmer=PorterStemmer()
print(" ".join(stemmer.stem(word) for word in vocabulary))
```

```python
#snowball stemmer
from nltk.stem.snowball import SnowballStemmer
stemm=SnowballStemmer(language='english')
print(" ".join([stemm.stem(word) for word in vocabulary]))
```

```python
# Lancaster stemmer
from nltk.stem.lancaster import LancasterStemmer
lstemmer=LancasterStemmer()
print(" ".join([lstemmer.stem(word) for word in vocabulary]))
```

# Cleaning the words

```python
#text in lower cases
new_text=article_text.lower()
new_text
```

```python
import re
new1_text=re.sub('^a-zA-Z',' ',new_text)
new1_text
```

```python
new_text2=re.sub(r'\s+',' ',new1_text)
new_text2
```

```python
sentences=nltk.sent_tokenize(new_text2)
```

```python
# Removing stopwords
from nltk.corpus import stopwords
new_vocab=[nltk.word_tokenize(sent) for sent in sentences]
for i in range(len(new_vocab)):
    new_vocab[i]=[w for w in new_vocab[i] if w not in stopwords.words('english')]
    
    
```
